# GENDER

```python
# Bar-plot that tells which gender is puchasing more goods on Diwali.
ax=sns.countplot(x='Gender',data=df,ec='black',lw=1)

for bars in ax.containers:
    ax.bar_label(bars)
plt.grid(True)
plt.title("PURCHASING",fontsize=15)
plt.xlabel('Gender of the customer',fontsize=13)
plt.ylabel('Number of Purchasing',fontsize=13)
plt.show()
```

From the above graph we can say that most of the buyers are Females.

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sales_gen
```

```python
# Bar-plot that tells which gender is spending more money on purchasing.
sns.barplot(x='Gender',y='Amount',data=sales_gen,ec='black',lw=1)
plt.grid(True)
plt.title("SPENDING",fontsize=15)
plt.xlabel('Gender of the customer',fontsize=13)
plt.ylabel('Amount of Purchasing',fontsize=13)
plt.show()
```

From the above graph we can say that females are spending more money on purchasing as comapred to male.

# AGE

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',ec='black',lw=1,hue='Gender')
plt.grid(True)
plt.title("PURCHASING",fontsize=15)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
Sales_age=df.groupby(["Age Group"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
Sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=Sales_age,ec='black',lw=1)
plt.title("SPENDING",fontsize=15)
plt.grid(True)
plt.show()
```

From the above graph we can say that the buyers between 26-35 age group are spending more money on buying goods on diwali.

# STATE

```python
# Total number of orders from top 10 states.
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(x='State',y='Orders',data=sales_state,ec='black',lw=1)
plt.title("PURCHASING",fontsize=15)
plt.grid(True)
plt.show()
```

From the above graph we can say that most of the orders are from Uttar Pradesh, Maharashtra and Karnataka.

```python
# Total amount/sales from top 10 states.
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(x='State',y='Amount',data=sales_state,ec='black',lw=1)
plt.title("SPENDING",fontsize=15)
plt.grid(True)
plt.show()
```

From the above graph we can say that Uttar Pradesh, Maharashtra and Karnataka are spending more money on buying goods in diwali.

# OCCUPATION

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df,ec='black',lw=1)
for bars in ax.containers:
    ax.bar_label(bars)
plt.grid(True)
plt.xlabel('Occupation',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.show()
```

# PRODUCT CATEGORY

```python
sns.set(rc={'figure.figsize':(30,10)})
ax=sns.countplot(data=df,x='Product_Category',ec='black',lw=1)
for bars in ax.containers:
    ax.bar_label(bars)
plt.grid(True)
plt.xlabel('Product_Category',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.show()
```

```python
product_category=df.groupby(['Product_Category'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False).head(10)
product_category
```

```python
sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='Product_Category',data=product_category,y='Amount',ec='black',lw=1)
plt.xlabel('Product_Category',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.grid(True)
plt.show()
```

From above graphs we can say that most of the sold products are from Food, Clothing and Electronics category.

# PRODUCT

```python
product=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
product
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=product,x='Product_ID',y='Orders',ec='black',lw=1)
plt.xlabel('Product_ID',fontsize=15)
plt.ylabel('Number of Orders',fontsize=15)
plt.grid(True)
plt.show()
```

```python
# Top 10 most sold products.
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar',y='Amount',ec='black',lw=1,color=['red','aqua','blue','brown','chartreuse','yellow','purple','pink','#FF1493','#FF00FF'])
plt.grid(True)
plt.xlabel('Product_ID',fontsize=15)
plt.ylabel('Number of Orders',fontsize=15)
plt.show()
```

From the above graph we can say that P00265242 and P00110942 are the most sold products.
